### Determina para qué valores de $x$ se puede sustituir $\cos x$ por $1 - x^{2} /2$ si el error absoluto máximo permitido es $10^{-4}$, utilizando la cota de error del polinomio de Taylor. Representa gráficamente el problema.


SOLUCIÓN:

Definimos la función $f(x) = \cos(x)$

```python
f(x)=cos(x)
f
```

Veamos que polinomio de Taylor de $\cos x$ se corresponde con $1 - x^{2} /2$. Hagamos los cuatros primeros órdenes.

```python
show(taylor(f(x),x,0,1));
show(taylor(f(x),x,0,2));
show(taylor(f(x),x,0,3));
show(taylor(f(x),x,0,4))
```

Tomamos el de orden 3, pues nos dará mejor aproximación.

```python
P(x)=taylor(f(x),x,0,3) 
show(P(x))
```

Veamos cómo es la aproximación intuitivamente usando plot.

```python
plot(f,-1.2,1.2)+plot(P,-1.2,1.2,color='red')
```

OBSERVACIÓN GENERAL:

Sabemos que la expresión del error para el polinomio de Taylor $P_n(x)=P_n(f,a)(x)$ de la función $f$ centrado en el punto $a$, de grado $\leq n$, viene dada por 

$$|f(x)-P_n(x)|\leq \frac{\|f^{(n+1)}\|_{[a,x]}}{(n+1)!}|x-a|^{n+1}$$

Por tanto, debemos estimar 

$$M_n=\|f^{(n+1)}\|_{[a,x]}$$ 

EN EL CASO EN EL QUE NOS COMPETE

En concreto, las $x$ están por determinar, $a=0$ y $n=3$

$$|f(x)-P_3(x)|\leq \frac{\|f^{(4)}\|_{[0,x]}}{4!}|x-0|^{4}$$
y estimamos

$$M_3=\|f^{(4)}\|_{[0,x]}$$

Como las sucesivas derivadas de $\cos(t)$ son $-\sin (t)$, $-\cos (t)$, $\sin(t)$ y $\cos(t)$ podemos concluir que $M_n=1$ para todo $n$. En particular, para nuestra $n=3$ lo es.

Por tanto, la cota es: 

```python
cota(x)=abs(x)^4/factorial(4)
cota(x)
```

Ahora vamos a ver qué valores de $x$ cumplen $cota(x)\leq 10^{-4}$:

$$\frac{|x|^4}{24}\leq 10^{-4} \Rightarrow |x|^4\leq 10^{-4}\cdot 24\Rightarrow |x|\leq \left(10^{-4}\cdot 24\right)^{\frac{1}{4}}$$

```python
r=(24*10^(-4))^(1/4).n()
print('x debe estar entre ',-r, 'y' , r)
```

La representación gráfica de la función $f(x)$ y el polinomio $P(x)$ en el intervalo $[-r- 0.5,r+0.5]$ indica que el polinomio aproxima muy bien la función, y podréis pensar que el valor de a podría ser mayor:

```python
plot(cos(x),(x,-r-.4,r+.4),legend_label='$y=cos(x)$')+plot(P,(x,- r-.4,r+.4),legend_label='$y=P(x)$',color='red')+point2d([[r,0]],color='black',size=50)+point2d([[-
 r,0]],color='black',size=50)
```

Pero podemos verlo mejor dibujando la función de "error" $|f(x) − P (x)|$ en dicho intervalo

```python
plot(abs(cos(x)-P(x)),(x,-r-0.2,r+0.2))+plot(10^(-4),(x,-r-0.2,r+0.2),color='red')+point2d([[r,0]],size=50)+point2d([[-r,0]],size=50)
```
