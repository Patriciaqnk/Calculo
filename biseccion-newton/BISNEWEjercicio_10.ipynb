### a) Encuentra un intervalo de anchura $0.2$ al que pertenezca $\alpha$ y en el que se cumplan las condiciones del “teorema de convergencia global” del método de Newton.

```python
a=0.8
b=1
```

Comprobamos que se cumplen las condiciones del teorema de convergencia global del método de Newton en ese intervalo.
<p>Vamos a ver que en el intervalo [a,b] se cumplen las condiciones del 

<strong>Teorema de Convergencia Global del Método de Newton</strong>:</p>
<p><br />$i)$ $f(a)f(b)<0$<br /><br />$ii)$ $f'(x)\neq 0\,\,\forall x\in [a,b]$<br /><br />$iii)$ $f''(x)$ no cambia de signo en $[a,b]$<br /><br />$iv)$ max$\left\{\left|{f(a)\over f'(a)}\right|,\left|{f(b)\over f'(b)}\right|\right\}\le b-a$</p>
<p> </p>
<p>entonces el método de Newton</p>
<p>$$ \left\{ \begin{array}{l} x_{0}\in [a,b] \\  x_{n+1}=x_n-\frac{f(x_{n})}{f'(x_n)}\end{array}\right.$$</p>
<p>converge  a la única solución $\alpha $ con $f(\alpha)=0$ en el intervalo $[a,b]$.</p>
<p> </p>
<p>Comprobémoslo:</p>
<p> </p>
<p> $i)$ $f(a)f(b)<0$</p>

```python
(f(a)*f(b)).n()
```

<p>$ii)$ $f'(x)\neq 0\,\,\forall x\in [a,b]$</p>
<p>Esto lo podemos observar con la gráfica de $y=f'(x)$. De todas formas, vemos que crece la función en el intervalo elegido, por lo que la derivada es creciente.</p>

```python
plot(diff(f(x),x),a,b) + plot(0, a,b,color='red')
```

```python
f;diff(f(x),x);diff(f(x),x,2);
```

<p>$iii)$ $f''(x)$ no cambia de signo en $[a,b]$.</p>

```python
plot(diff(f(x),x,2),(x,a,b))+ plot(0,(x,a,b),color='red')
```

NO se cumplen las condiciones. Por tanto cambiamos el intervalo.

### b)  Realiza manualmente dos pasos del método de Newton empezando en un punto del intervalo obtenido en el apartado (a).

Realizamos a mano dos pasos:

```python
x0=(a+b)/2
x1=x0-f(x0)/d1(x0)
x2=x1-f(x1)/d1(x1)
[x0,x1,x2] 
#aproximaciones sucesivas dadas or los tres primeros términos de la sucesión
```

### c) Calcula aproximadamente $\alpha$  mediante el método de Newton, deteniendo el algoritmo cuando la distancia entre dos iteraciones consecutivas sea menor que $10^{−8}$.

```python
x0=(a+b)/2
x1=x0-f(x0)/d1(x0)
x2=x1-f(x1)/d1(x1)
x3=x2-f(x2)/d1(x2)
abs(x3-x2)
#Observamos que aún no es suficiente por lo que iteramos una vez más.
```

```python
x0=(a+b)/2
x1=x0-f(x0)/d1(x0)
x2=x1-f(x1)/d1(x1)
x3=x2-f(x2)/d1(x2)
x4=x3-f(x3)/d1(x3)
abs(x4-x3)
```

Por lo tanto, el valor aproximado de $\alpha$ es:

```python
x4
```
