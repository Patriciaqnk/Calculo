### a) ¿Cuántas soluciones tiene menores que 10 en valor absoluto?

Vamos a aplicar el método de Newton para aproximar la raíz de $(x-1)\sin(x)=\dfrac{1}{2(1-x^2)}$ 

```python
g(x)=(x-1)*sin(x)-1/(2*(1-x^2))
g
```

¿Cuántas raíces hay? 

```python
plot(g,-5,5)#sin pensar
```

Trabajamos la ecuación en primer lugar,

$$(x-1)\sin(x)=\dfrac{1}{2(1-x^2)}\rightarrow 1 = (x-1)\sin(x)2(1-x^2) \rightarrow 1-(x-1)\sin(x)2(1-x^2)=0$$

Así podemos tomar como 
$$f(x)= 1-(x-1)\sin(x)2(1-x^2) =$$
$$ = 1+(1-x)\sin(x)2(1-x^2) = $$
$$ = 1 + 2(x^{3}-x^{2}-x+1)\sin(x)$$

o bien, usar la sentencia "numerator(g)".

```python
show(1+2*(x^3-x^2-x+1)*sin(x))
```

```python
f=numerator(g)
show(f)
```

```python
plot(f,-10,10,color='green')
# Al pedir las soluciones menores que 10 en valor absoluto buscamos en [-10,10]
```

```python
plot(f,-4,4,color='green')
#En el anterior gráfico se distinguen 4 pero si nos acercamos observamos 2 más.
```

```python
plot(f,-1,0,color='green')
#Y si nos volvemos a acercar aparecen otras dos.
```

Con lo cual hay 8 raíces reales menores que 10 en valor absoluto.
