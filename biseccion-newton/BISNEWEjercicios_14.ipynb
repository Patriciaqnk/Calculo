### b) Realiza manualmente dos pasos del método de Newton empezando en $x_0 = 2$. ¿Qué ocurre?

Hagamos a mano dos pasos del método de Newton empezando en $x_0=2$. 
Recordamos que la función recursiva del método de Newton es $$x_{n+1}=x_n-f(x_n)/f'(x_n)$$

```python
d1=diff(f,x)
d2=diff(f,x,2)
show(d1)
show(d2)
```

```python
x0=2 
x1=x0-f(x0)/d1(x0)
x2=x1-f(x1)/d1(x1)
show(x2.n())
```

<p>Vemos que nos hemos alejando de las soluciones cercanas al 2. Esto es porque alrededor del 2, gráficamente vemos un mínimo, es decir, la derivada se anula. Además la recta tangente en el x=2 tiene muy poca pendiente y su intersección con la función está muy lejos del x=2. Observamos también que tras dos pasos nos aproximamos a la raíz que hay cerca del punto x=-98,  aunque hayamos empezado mucho más cerca de las raíces positivas.</p>

```python
plot(f,0.5,3)+plot(d1(2)*(x-x0)+f(2),0.5,3,color='red')
#dibujo la función en un intervalo donde está el 2 y las soluciones cercanas a 2, 
#y además dibujamos la recta tangente que pasa por el punto (2,f(2)).
```

### (c) Sea $\alpha$ la solución negativa de la ecuación. Encuentra un intervalo de anchura 0.5 al que pertenezca $\alpha$ y en el que se cumplan las condiciones del “Teorema de convergencia global” del método de Newton.

Busquemos un intervalo alrededor de la solución negativa que verifique Newton.

```python
b=-97.75
a=-98.25
b-a
```

<p><strong> </strong>Vamos a ver que en el intervalo $[a,b]$ se cumplen las condiciones del <strong>"teorema de convergencia global" del método de Newton</strong>:</p>
<p><br /><span style="background-color: #ffff99;">Sea $f:[a,b]\rightarrow {\mathbb R}$ de clase $C^{2}$ en $[a,b]$, cumpliendo las condiciones:</span><br /><br /><span style="background-color: #ffff99;">$i)$  $f(a)f(b)<0$.</span><br /><br /><span style="background-color: #ffff99;">$ii)$  $f'(x)\neq 0\,\,\forall x\in [a,b]$.</span><br /><br /><span style="background-color: #ffff99;">$iii)$  $f''(x)$ no cambia de signo en $[a,b]$.</span><br /><br /><span style="background-color: #ffff99;">$iv)$  max$\left\{\left|{f(a)\over f'(a)}\right|,\left|{f(b)\over f'(b)}\right|\right\}\le b-a$.</span></p>
<p><span style="background-color: #ffff99;">Entonces el método de Newton converge  a la única solución de la ecuación $f(x)=0$ en el intervalo $[a,b]$.</span></p>
<p> </p>
<p>Comprobémoslo:</p>
<p> </p>
<p> $i)$  $f(a)f(b)<0$.</p>

```python
f(a)*f(b)
```

<p>$ii)$  $f'(x)\neq 0\,\,\forall x\in [a,b]$.</p>
<p>Esto lo podemos observar con la gráfica de $y=f'(x)$: la derivada nunca se anula.</p>

```python
plot(diff(f(x),x),(x,a,b)) + plot(0,(x,a,b),color='red')
```

También podemos obervar con la gráfica de $f''(x)$ que la derivada segunda es negativa en dicho intervalo:

```python
plot(diff(f(x),x,2),(x,a,b)) + plot(0,(x,a,b),color='red')
```

<p>$iv)$  max$\left\{\left|{f(a)\over f'(a)}\right|,\left|{f(b)\over f'(b)}\right|\right\}\le b-a$.</p>

```python
b-a
```

<p>El  max$\left\{\left|{f(a)\over f'(a)}\right|,\left|{f(b)\over f'(b)}\right|\right\} $ es</p>

### (d) Calcula aproximadamente la raíz mediante el método de Newton, deteniendo el algoritmo cuando la distancia entre iteraciones consecutivas sea menor que $10^{-10}$.

```python
eps=10^(-10)
```

```python
contador =1
N=50 # Nº máximo de iteraciones (por si el criterio de parada no se detiene)
x0=(a+b)/2
x1=x0-f(x0)/d1(x0)
while abs(x0-x1)>eps and contador < 50:
    x0=x1
    x1=x0-f(x0)/d1(x0)
    contador=contador + 1
print('La solución aproximada es', x1)
```
